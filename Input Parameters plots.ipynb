Theory Behind-
Normal distribution, also called Gaussian distribution, the most common distribution function for independent, randomly generated variables. Its familiar bell-shaped curve is ubiquitous in statistical reports, from survey analysis and quality control to resource allocation.

The graph of the normal distribution is characterized by two parameters: the mean, or average, which is the maximum of the graph and about which the graph is always symmetric; and the standard deviation, which determines the amount of dispersion away from the mean. A small standard deviation (compared with the mean) produces a steep graph, whereas a large standard deviation (again compared with the mean) produces a flat graph.The normal distribution is produced by the normal density function, p(x) = e−(x − μ)2/2σ2/σSquare root of√2π.

```python
#NORMAL DISTRIBUTION FOR X-COMPOSITION
import matplotlib.pyplot as plt
import scipy.stats
import numpy as np
x_min = 0.287
x_max = 0.293
mean = 0.2900 
#std = 0.00016666666
std=0.0005 # that is 3*std=0.0005
x = np.linspace(x_min, x_max, 1000)
y = scipy.stats.norm.pdf(x,mean,std)
plt.plot(x,y, color='red')
plt.grid()
plt.xlim(x_min,x_max)
#plt.ylim(0,12000)
plt.title('Normal Distribution For x-composition',fontsize=25)
plt.xlabel('x-Composition',fontsize=15)
plt.ylabel('Number Of Elements',fontsize=15)
plt.fill_between(x,y, 0,alpha = 0.2, color = 'purple')
# plotting mean line
plt.axvline(mean, color='k', linestyle='dashed', linewidth=2)
plt.show()
```

```python
#NORMAL DISTRIBUTION FOR NA
import matplotlib.pyplot as plt
import scipy.stats
import numpy as np
x_min =0.2e15
x_max =1.4e16
mean = 7.5e15  #cm^-3
std=8.3333333e+14  #cm^-3
x = np.linspace(x_min, x_max, 1000)
y = scipy.stats.norm.pdf(x,mean,std)
plt.plot(x,y, color='red')
plt.grid()
plt.xlim(x_min,x_max)
plt.title('Normal Distribution For NA(cm^-3)',fontsize=25)
plt.xlabel('NA',fontsize=15)
plt.ylabel('Number Of Elements',fontsize=15)
plt.fill_between(x,y, 0,alpha = 0.2, color = 'purple')
# plotting mean line
plt.axvline(mean, color='k', linestyle='dashed', linewidth=2)
plt.show()
```
