For monte Carlo simulations we need to check whether any correlation exists between input parameters
Two ways to check
1)To calculate the correlation between two variables in Python, we can use the Numpy corrcoef() function.Value zero means no correlation
2)plot the graph taking one variable on x axis and other on y axis and check it must be random one 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
var = pd.read_excel("/content/check1 _excel.xlsx")
var.head()
```
